```python
import pandas as pd
#import plotly_express as px

import folium
from folium.plugins import HeatMap
```

```python
df = pd.read_csv("./formatted.csv")
df
```

```python
m = folium.Map(location=[df.Latitude.mean(), df.Longitude.mean()], zoom_start=8, control_scale=True)
```

```python
map_values = df[['Latitude', 'Longitude']]

data = map_values.values.tolist()
```

```python
hm = HeatMap(data, min_opacity=0.05, max_opactiy=0.9, radius=25).add_to(m)
m
```
